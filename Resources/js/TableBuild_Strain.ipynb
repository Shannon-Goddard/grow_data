```python
# Import dependencies
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import os 
from urllib.request import urlopen
import re

#initiate data storage
page = 1
titles = []
about = []
while page != 128:

    # scraping wikileaf website
    url_link = f'https://www.wikileaf.com/strains/?page={page}'
    
    request = requests.get(url_link)
 
    Soup = BeautifulSoup(request.text, "html.parser")
 
    # creating a list of all common heading tags
    heading_tags = ["h5"]
    for tags in Soup.find_all(heading_tags, class_='name disp-title'):
        
        #name
        name = tags.text.strip()
        titles.append(name)
        
        #about
        abouts = f'https://www.wikileaf.com/strain/{name}/'
        about.append(abouts)
        
    page = page + 1
```

```python
#pandas dataframe        
strains = pd.DataFrame({
'strain': titles,
'about' : about,
})

#df.columns = df.columns.str.replace(' ','_')
strains.about = strains.about.replace(regex=[' '], value='-')
strains.about = strains.about.str.lower()  
# displaying the DataFrame
display(strains.head(10))
```

```python
# Save as .csv
os.makedirs('../Resources/', exist_ok=True)  
strains.to_csv('../Resources/strain_data.csv') 
```
